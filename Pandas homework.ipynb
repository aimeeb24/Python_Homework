```python
year_counts = lego_df['year_released'].value_counts()                                                      # Chat GPT used to get the code - I added colour and changed figsize
year_counts = year_counts.sort_index()

plt.figure(figsize=(10, 4))  
plt.bar(year_counts.index, year_counts.values, width=0.7, align='center', color='gold')

plt.xlabel('Year Released')
plt.ylabel('Number of Sets')
plt.title('Lego Sets Released per Year')

plt.tight_layout()  # Ensure labels and title fit within the figure
plt.show()
```

```python
filtered_lego_df = lego_df[lego_df['number_of_parts'] > 5000]                                                    # Chat GPT used to get code - with edits (colour map)
                                                                                                 # Tried multiple ways to get rid of the decimals on the years - nothing works
# Create a color map for unique sets
unique_sets = filtered_lego_df['set_name'].unique()
num_unique_sets = len(unique_sets)
colors = plt.cm.Spectral(np.linspace(0, 1, num_unique_sets))

# Create a scatter plot for each set with a unique color
plt.figure(figsize=(10, 6)) 

for i, set_name in enumerate(unique_sets):
    subset = filtered_lego_df[filtered_lego_df['set_name'] == set_name]
    plt.scatter(subset['number_of_parts'], subset['year_released'], label=set_name, alpha=0.5, color=colors[i])

# Add labels and a title
plt.xlabel('Number of Parts')
plt.ylabel('Year Released')
plt.title('Scatter Plot of Lego Sets with More Than 5000 Parts')

# Create a separate legend at the side
plt.legend(title='Set Names', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True)  
plt.tight_layout() 
plt.show()
```

```python
star_wars_df = lego_df[lego_df['theme_name'] == 'Star Wars']                                    # chat gpt with edits (colour, linestyle, deleted marker)

# Group the data by 'year_released' and count the number of sets released each year
yearly_counts = star_wars_df.groupby('year_released').size()


plt.figure(figsize=(10, 4))
plt.plot(yearly_counts.index, yearly_counts.values, linestyle='dotted', color='purple')

plt.xlabel('Year Released')
plt.ylabel('Number of Star Wars Sets')
plt.title('Star Wars Lego Sets Released Over the Years')

plt.tight_layout()
plt.show()
```

```python
star_wars_df = lego_df[lego_df['theme_name'] == 'Star Wars']                                                     # adding another line to compare sets released
harry_potter_df = lego_df[lego_df['theme_name'] == 'Harry Potter']                                               # used chat gpt to fix code 

star_wars_counts = star_wars_df.groupby('year_released').size()
harry_potter_counts = harry_potter_df.groupby('year_released').size()

plt.figure(figsize=(10, 4)) 
plt.plot(star_wars_counts.index, star_wars_counts.values, linestyle='dotted', color='purple', label='Star Wars')
plt.plot(harry_potter_counts.index, harry_potter_counts.values, linestyle='dotted', color='orange', label='Harry Potter')


plt.xlabel('Year Released')
plt.ylabel('Number of Sets')
plt.title('Lego Sets Released Over the Years (Star Wars vs. Harry Potter)', fontweight='bold')

plt.legend()

plt.tight_layout()
plt.show()
```
